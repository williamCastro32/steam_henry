```python
# se importan las librerias que se van a usar 
import gzip
import ast
import chardet
import pandas as pd
import nltk
```

Primero se procede a abrir el archivo user_items

```python
def descomprimir_file_items(ruta, variable_anidada):
    # Detección automática del encoding
    with open(ruta, 'rb') as archivo_binario:
        resultado = chardet.detect(archivo_binario.read())
        encoding_detectado = resultado['encoding']

    # Lectura del archivo utilizando el encoding detectado
    fila = []
    with gzip.open(ruta, 'rt', encoding=encoding_detectado) as archivo:
        for line in archivo.readlines():
            fila.append(ast.literal_eval(line))

    # Creación del DataFrame y procesamiento adicional
    df = pd.DataFrame(fila)
    df = df.explode(variable_anidada).reset_index(drop=True)

    if not df.empty:
        df = pd.concat([df, pd.json_normalize(df[variable_anidada])], axis=1)
        df = df.drop(columns=variable_anidada)

    return df

def descomprimir_file_items(ruta, variable_anidada):
    '''Esta función recibe una ruta de acceso a un archivo json anidado y carga la información en un
    DataFrame'''
    fila = []
    with gzip.open(ruta, 'rt', encoding='MacRoman') as archivo:
        for line in archivo.readlines():
            fila.append(ast.literal_eval(line))

    df = pd.DataFrame(fila)
    df = df.explode(variable_anidada).reset_index(drop=True)
    
    if not df.empty:
        df = pd.concat([df, pd.json_normalize(df[variable_anidada])], axis=1)
        df = df.drop(columns=variable_anidada)

    return df
```

```python
file_path_items = "../datasets/users_items.json.gz"
```

```python
user_items = descomprimir_file_items(file_path_items,'items')
```
