```python
import gzip
import json
import pandas as pd

nombre_archivo = 'steam_games.json.gz'

# Lista para almacenar los datos
datos_lista = []

# Abrir el archivo comprimido en modo lectura binaria
with gzip.open(nombre_archivo, 'rt', encoding='utf-8') as archivo:
    # Leer cada línea del archivo y cargar el JSON
    for linea in archivo:
        try:
            datos = json.loads(linea)
            datos_lista.append(datos)
        except json.JSONDecodeError as e:
            print(f"Error al decodificar JSON: {e}")

# Convertir la lista de datos a un DataFrame
df = pd.DataFrame(datos_lista)

# Ahora 'df' es un DataFrame de pandas que puedes utilizar
# para realizar análisis y manipulación de datos
print(df.head())
```
